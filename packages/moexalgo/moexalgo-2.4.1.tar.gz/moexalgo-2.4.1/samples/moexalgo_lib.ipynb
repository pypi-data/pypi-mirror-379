## 3. Можно получить tradelog за сегодня  

````python
# лимит на выдачу данных: 10_000 записей
lkoh = Ticker('LKOH')
lkoh.trades()
````

````python
# следующую порцию данных можно получить передав параметр 'tradeno' 
# получить сделки, у которых tradeno >= 11590030639
lkoh.trades(tradeno=11590030639)
````

````python
rim4 = Ticker('RIM4')
rim4.candles(start='2024-01-01', end='2024-12-31', period='1d')
````
